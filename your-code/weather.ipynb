## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
from sqlalchemy import create_engine, text
import pandas as pd
import numpy as np
```

```python
#data = pd.read_csv('~/Documents/Ironhack/Lab/lab-data-cleaning/weather-raw.csv')
data = pd.read_csv('~/ironhackt/module1/labs/lab-data-cleaning/weather-raw.csv')
```

```python
data.head()
```

```python
formatted_df = pd.melt(data,
                       ["id","year","month","element"],
                       var_name="dia",
                       value_name="tem")
formatted_df = formatted_df.sort_values(by=["month","dia"])
formatted_df.head(10)
```

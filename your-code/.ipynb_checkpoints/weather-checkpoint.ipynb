## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
from sqlalchemy import create_engine, text
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('~/Documents/Ironhack/Lab/lab-data-cleaning/weather-raw.csv')
```

```python
data.head()
```

```python
formatted_df = pd.melt(data,
                       ["id","year","month","element"],
                       var_name="dia",
                       value_name="tem")
formatted_df = formatted_df.sort_values(by=["month","dia"])
formatted_df.head(10)
```

```python
lista_nulos = formatted_df.isnull().sum(axis = 1)
lista_nulos
```

```python
lista_nulos > 0
```

```python
total_nulos_2 = formatted_df.isnull().T
total_nulos_2
```

```python
formatted_df.shape
```

```python
lista_nulos[lista_nulos > 0]
```

```python
lista_nulos_colum = lista_nulos[lista_nulos > 0].index
lista_nulos_colum
```

```python
formatted_df.T
```

```python
lista_limpia = formatted_df.T.drop(columns = lista_nulos_colum)
lista_limpia
```

```python
lista_limpia = lista_limpia.T
lista_limpia
```

```python
tabla_med = lista_limpia[['month','element','tem']]
tabla_med
```

```python
formatted_df_tem = pd.melt(tabla_med,
                       ["month"],
                       var_name="tem",
                       value_name="tem2")
formatted_df_tem = formatted_df_tem.sort_values(by=["month"])
formatted_df_tem.head(10)
```

```python
df_unmelted_t = tabla_med.pivot(index='month', columns='tem')
df_unmelted_t
```
